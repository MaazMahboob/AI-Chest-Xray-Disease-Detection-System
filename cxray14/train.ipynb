```python

import streamlit as st
import sqlite3, os, datetime, json, hashlib, io, re
import numpy as np
import cv2
import requests
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from PIL import Image
from ultralytics import YOLO
from fpdf import FPDF

# ================================================================
# CONFIG
# ================================================================
MODEL_PATH = "cxray14/runs/detect/runs/cxr14_yolov12m/weights/best.pt"
UPLOAD_DIR   = "uploads"
DB_PATH      = "database/scans.db"
OLLAMA_URL   = "http://localhost:11434/api/generate"
OLLAMA_MODEL = "qwen3:14b"   # or deepseek-r1:14b

CLASS_NAMES = [
    "Aortic enlargement","Atelectasis","Calcification","Cardiomegaly",
    "Consolidation","ILD","Infiltration","Lung Opacity",
    "Nodule/Mass","Other lesion","Pleural effusion","Pleural thickening",
    "Pneumothorax","Pulmonary fibrosis"
]

SEVERITY = {
    "Pneumothorax":"CRITICAL",    "Cardiomegaly":"HIGH",
    "Pleural effusion":"HIGH",    "Consolidation":"HIGH",
    "Lung Opacity":"HIGH",        "Nodule/Mass":"HIGH",
    "Pulmonary fibrosis":"MODERATE","ILD":"MODERATE",
    "Aortic enlargement":"MODERATE","Infiltration":"MODERATE",
    "Atelectasis":"MODERATE",     "Pleural thickening":"LOW",
    "Calcification":"LOW",        "Other lesion":"LOW",
}

SEV_COLOR  = {"CRITICAL":(220,20,60),"HIGH":(255,100,0),"MODERATE":(220,180,0),"LOW":(0,200,100)}
SEV_ORDER  = {"CRITICAL":0,"HIGH":1,"MODERATE":2,"LOW":3}

os.makedirs(UPLOAD_DIR, exist_ok=True)
os.makedirs("database",  exist_ok=True)

# ================================================================
# PAGE CONFIG + CSS
# ================================================================
st.set_page_config(page_title="Chest X-ray AI", page_icon="🫁",
                   layout="wide", initial_sidebar_state="expanded")

st.markdown("""
<style>
[data-testid="stAppViewContainer"]{background:#0f1117;color:#e0e0e0}
[data-testid="stSidebar"]{background:#1a1d27}
.card{background:#1e2130;border-radius:12px;padding:18px 22px;
margin:8px 0;border-left:4px solid #4f8ef7}
.finding-row{background:#1e2130;border-radius:8px;padding:10px 16px;
margin:5px 0;display:flex;justify-content:space-between;align-items:center}
.sev-CRITICAL{color:#dc143c;font-weight:700}
.sev-HIGH{color:#ff6400;font-weight:700}
.sev-MODERATE{color:#ddb200;font-weight:700}
.sev-LOW{color:#00c864;font-weight:700}
h1,h2,h3{color:#ffffff!important}
hr{border-color:#2e3250}
</style>
""", unsafe_allow_html=True)

# ================================================================
# DATABASE
# ================================================================
@st.cache_resource
def get_db():
    conn = sqlite3.connect(DB_PATH, check_same_thread=False)
    conn.execute("""CREATE TABLE IF NOT EXISTS users(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        username TEXT UNIQUE, password TEXT, role TEXT)"""
    )
    conn.execute("""CREATE TABLE IF NOT EXISTS scans(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        doctor TEXT, patient TEXT, filename TEXT,
        findings TEXT, report TEXT, date TEXT, conf REAL)"""
    )
    conn.commit()
    return conn

db = get_db()

def hp(pw): return hashlib.sha256(pw.encode()).hexdigest()

def register(u, p, role):
    try:
        db.execute("INSERT INTO users(username,password,role) VALUES(?,?,?)", (u,hp(p),role))
        db.commit(); return True
    except: return False

def login(u, p):
    row = db.execute("SELECT role FROM users WHERE username=? AND password=?",
                     (u,hp(p))).fetchone()
    return row[0] if row else None

# ================================================================
# MODEL
# ================================================================
@st.cache_resource
def load_model():
    return YOLO(MODEL_PATH) if os.path.exists(MODEL_PATH) else None

model = load_model()

# ================================================================
# OLLAMA REPORT
# ================================================================
def ollama_report(patient, findings):
    if not findings:
        return "No significant radiological findings detected. Lung fields appear clear bilaterally."
    lines = "\n".join(
        f"- {f['label']} (confidence {f['conf']:.0%}, severity {f['severity']})"
        for f in findings
    )
    prompt = (
        f"You are a senior radiologist. Write a concise clinical impression (3-4 sentences, "
        f"professional prose, no bullet points) for a chest X-ray.\n\n"
        f"Patient: {patient}\nAI-detected findings:\n{lines}\n\n"
        f"Write the clinical impression only."
    )
    try:
        r = requests.post(OLLAMA_URL, json={
            "model": OLLAMA_MODEL, "prompt": prompt,
            "stream": False, "options": {"temperature":0.3,"num_predict":350}
        }, timeout=90)
        text = r.json().get("response","").strip()
        text = re.sub(r"<think>.*?</think>", "", text, flags=re.DOTALL).strip()
        return text or "Report generation returned empty response."
    except Exception as e:
        return (f"[Ollama offline — run 'ollama serve' in terminal]\n"
                f"Findings: {', '.join(f['label'] for f in findings)}.")

# ================================================================
# DETECTION
# ================================================================
def detect(image, conf_thresh):
    img = np.array(image.convert("RGB"))
    res = model.predict(img, conf=conf_thresh, verbose=False)
    boxes   = res[0].boxes.xyxy.cpu().numpy()
    classes = res[0].boxes.cls.cpu().numpy()
    scores  = res[0].boxes.conf.cpu().numpy()

    findings = []
    draw = img.copy()

    for box, cls, score in zip(boxes, classes, scores):
        x1,y1,x2,y2 = map(int, box)
        label  = CLASS_NAMES[int(cls)]
        sev    = SEVERITY.get(label, "LOW")
        color  = SEV_COLOR[sev]
        findings.append({"label":label,"conf":float(score),"severity":sev,"box":[x1,y1,x2,y2]})
        cv2.rectangle(draw,(x1,y1),(x2,y2),color,2)
        txt = f"{label} {score:.0%}"
        (tw,th),_ = cv2.getTextSize(txt,cv2.FONT_HERSHEY_SIMPLEX,0.55,1)
        cv2.rectangle(draw,(x1,y1-th-8),(x1+tw+4,y1),color,-1)
        cv2.putText(draw,txt,(x1+2,y1-4),cv2.FONT_HERSHEY_SIMPLEX,0.55,(255,255,255),1)

    findings.sort(key=lambda x: SEV_ORDER.get(x["severity"],4))
    return Image.fromarray(draw), findings

# ================================================================
# PDF GENERATOR
# ================================================================
def make_pdf(patient, doctor, findings, report, date):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_fill_color(15,17,23)
    pdf.set_text_color(255,255,255)
    pdf.set_font("Helvetica","B",18)
    pdf.cell(0,14,"Chest X-ray AI Detection Report",ln=True,align="C",fill=True)

    pdf.ln(4)
    pdf.set_text_color(40,40,40)
    pdf.set_font("Helvetica","",11)
    pdf.cell(0,8,f"Patient: {patient}   |   Doctor: {doctor}   |   Date: {date}",ln=True)

    pdf.ln(4)
    pdf.set_font("Helvetica","B",13)
    pdf.cell(0,10,"Detected Findings:",ln=True)

    pdf.set_font("Helvetica","",11)
    clr = {"CRITICAL":(220,20,60),"HIGH":(200,80,0),"MODERATE":(180,140,0),"LOW":(0,150,80)}

    if findings:
        for f in findings:
            r,g,b = clr.get(f["severity"],(0,0,0))
            pdf.set_text_color(r,g,b)
            pdf.cell(0,8,f"  - {f['label']}  |  Conf: {f['conf']:.0%}  |  {f['severity']}",ln=True)
    else:
        pdf.set_text_color(0,150,80)
        pdf.cell(0,8,"  No abnormalities detected.",ln=True)

    pdf.ln(6)
    pdf.set_text_color(40,40,40)
    pdf.set_font("Helvetica","B",13)
    pdf.cell(0,10,"Clinical Impression (AI-generated):",ln=True)

    pdf.set_font("Helvetica","",11)
    pdf.multi_cell(0,8,report)

    pdf.ln(8)
    pdf.set_font("Helvetica","I",9)
    pdf.set_text_color(120,120,120)
    pdf.cell(0,8,"This report is AI-generated. Review by a licensed radiologist is required.",ln=True)

    buf = io.BytesIO()
    buf.write(pdf.output(dest="S").encode("latin-1"))  # FIXED LINE
    buf.seek(0)

    return buf

# ================================================================
# SESSION STATE
# ================================================================
if "logged_in" not in st.session_state:
    st.session_state.update({"logged_in":False,"username":"","role":""})

# ================================================================
# LOGIN / REGISTER
# ================================================================
if not st.session_state.logged_in:
    st.markdown("## 🫁 Chest X-ray AI Detection System")
    st.markdown("---")
    tab1, tab2 = st.tabs(["🔐 Login", "📝 Register"])

    with tab1:
        _, c, _ = st.columns([1,2,1])
        with c:
            u = st.text_input("Username", key="lu")
            p = st.text_input("Password", type="password", key="lp")
            if st.button("Login", use_container_width=True):
                role = login(u, p)
                if role:
                    st.session_state.update({"logged_in":True,"username":u,"role":role})
                    st.rerun()
                else:
                    st.error("Invalid credentials")

    with tab2:
        _, c, _ = st.columns([1,2,1])
        with c:
            u  = st.text_input("Username", key="ru")
            p  = st.text_input("Password", type="password", key="rp")
            rl = st.selectbox("Role", ["Doctor","Patient"])
            if st.button("Register", use_container_width=True):
                if register(u, p, rl): st.success("Registered! Please login.")
                else: st.error("Username already exists.")
    st.stop()

# ================================================================
# SIDEBAR
# ================================================================
uname = st.session_state.username
role  = st.session_state.role

with st.sidebar:
    st.markdown(f"### 👤 {uname}")
    st.caption(f"Role: **{role}**")
    st.markdown("---")

    if role == "Doctor":
        nav = st.radio("Navigation",
                       ["🔬 Run Detection","📋 Patient History","📊 Dashboard"])
    else:
        nav = "📋 Patient History"
        st.info("View your scan reports below.")

    st.markdown("---")
    try:
        requests.get("http://localhost:11434", timeout=2)
        st.success(f"🟢 Ollama Online ({OLLAMA_MODEL})")
    except:
        st.warning("🔴 Ollama Offline")
        st.caption("Open a terminal and run:\n`ollama serve`")
    st.markdown("---")
    if st.button("🚪 Logout"):
        st.session_state.logged_in = False; st.rerun()

# ================================================================
# PAGE: RUN DETECTION
# ================================================================
if nav == "🔬 Run Detection":
    st.markdown("# 🔬 X-ray Detection")

    if model is None:
        st.error("Model not found at model/best.pt — please copy best.pt into the model/ folder.")
        st.stop()

    left, right = st.columns(2)

    with left:
        patient = st.text_input("Patient Name", placeholder="e.g. Ahmed Khan")
        conf    = st.slider("Confidence Threshold", 0.10, 0.90, 0.40, 0.05)
        upload  = st.file_uploader("Upload Chest X-ray", type=["png","jpg","jpeg"])
        if upload:
            orig = Image.open(upload)
            st.image(orig, caption="Original", use_container_width=True)

    with right:
        if upload and st.button("🚀 Analyse X-ray", use_container_width=True):
            if not patient.strip():
                st.warning("Enter patient name first."); st.stop()

            with st.spinner("Running YOLO detection..."):
                ann, findings = detect(orig, conf)

            st.image(ann, caption="Detection Result", use_container_width=True)

            st.markdown("### Findings")
            if findings:
                for f in findings:
                    sev = f["severity"]
                    st.markdown(
                        f"<div class='finding-row'>"
                        f"<span>🔹 <b>{f['label']}</b></span>"
                        f"<span>{f['conf']:.0%}</span>"
                        f"<span class='sev-{sev}'>{sev}</span></div>",
                        unsafe_allow_html=True
                    )
            else:
                st.success("✅ No abnormalities detected")

            st.markdown("### 🤖 Clinical Impression")
            with st.spinner(f"Generating report with {OLLAMA_MODEL}..."):
                report = ollama_report(patient, findings)
            st.info(report)

            # Save annotated image
            ts    = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
            fname = f"{patient.replace(' ','_')}_{ts}.png"
            ann.save(os.path.join(UPLOAD_DIR, fname))

            db.execute(
                "INSERT INTO scans(doctor,patient,filename,findings,report,date,conf) VALUES(?,?,?,?,?,?,?)",
                (uname, patient, fname, json.dumps(findings), report,
                 str(datetime.datetime.now()), conf)
            )
            db.commit()
            st.success("💾 Saved to database")

            pdf_buf = make_pdf(patient, uname, findings, report,
                               datetime.datetime.now().strftime("%Y-%m-%d %H:%M"))
            st.download_button(
                "📥 Download PDF Report", data=pdf_buf,
                file_name=f"{patient}_report_{ts}.pdf",
                mime="application/pdf", use_container_width=True
            )

# ================================================================
# PAGE: PATIENT HISTORY
# ================================================================
elif nav == "📋 Patient History":
    st.markdown("# 📋 Scan History")

    if role == "Doctor":
        search = st.text_input("Search patient", placeholder="Leave blank for all")
        q = "SELECT doctor,patient,filename,findings,report,date FROM scans"
        p = []
        if search.strip():
            q += " WHERE patient LIKE ?"; p = [f"%{search}%"]
        q += " ORDER BY id DESC"
    else:
        q = "SELECT doctor,patient,filename,findings,report,date FROM scans WHERE patient=? ORDER BY id DESC"
        p = [uname]

    rows = db.execute(q, p).fetchall()

    if not rows:
        st.info("No records found.")
    else:
        for doc, pat, fname, fjson, rep, date in rows:
            findings = json.loads(fjson) if fjson else []
            with st.expander(f"👤 {pat}  —  {date[:16]}  |  Dr. {doc}"):
                ca, cb = st.columns(2)
                with ca:
                    ip = os.path.join(UPLOAD_DIR, fname)
                    if os.path.exists(ip):
                        st.image(ip, use_container_width=True)
                with cb:
                    st.markdown("**Findings:**")
                    if findings:
                        for f in findings:
                            sev = f.get("severity","LOW")
                            st.markdown(
                                f"<span class='sev-{sev}'>{sev}</span> — {f['label']} ({f['conf']:.0%})",
                                unsafe_allow_html=True
                            )
                    else:
                        st.write("No findings")
                    st.markdown("**Report:**")
                    st.info(rep)
                    pdf_buf = make_pdf(pat, doc, findings, rep, date[:16])
                    st.download_button("📥 Download PDF", data=pdf_buf,
                                       file_name=f"{pat}_report.pdf",
                                       mime="application/pdf", key=f"dl_{fname}")

# ================================================================
# PAGE: DASHBOARD
# ================================================================
elif nav == "📊 Dashboard":
    st.markdown("# 📊 Analytics Dashboard")

    rows = db.execute("SELECT patient,findings,date FROM scans ORDER BY id DESC").fetchall()

    if not rows:
        st.info("No data yet — run some detections first.")
    else:
        all_f = []
        for _, fjson, _ in rows:
            try: all_f.extend(json.loads(fjson))
            except: pass

        c1,c2,c3,c4 = st.columns(4)
        c1.metric("Total Scans",   len(rows))
        c2.metric("Patients",      len(set(r[0] for r in rows)))
        c3.metric("Total Findings",len(all_f))
        c4.metric("Critical Cases",sum(1 for f in all_f if f.get("severity")=="CRITICAL"))

        st.markdown("---")
        col1, col2 = st.columns(2)

        with col1:
            cnt = {}
            for f in all_f: cnt[f.get("label","?")] = cnt.get(f.get("label","?"),0)+1
            if cnt:
                df  = pd.DataFrame(cnt.items(), columns=["Finding","Count"]).sort_values("Count",ascending=False)
                fig = px.bar(df, x="Count", y="Finding", orientation="h",
                             color="Count", color_continuous_scale="Reds",
                             title="Finding Frequency")
                fig.update_layout(paper_bgcolor="#1e2130",plot_bgcolor="#1e2130",
                                  font_color="white",height=420)
                st.plotly_chart(fig, use_container_width=True)

        with col2:
            sc = {"CRITICAL":0,"HIGH":0,"MODERATE":0,"LOW":0}
            for f in all_f: sc[f.get("severity","LOW")] = sc.get(f.get("severity","LOW"),0)+1
            fig2 = go.Figure(go.Pie(
                labels=list(sc.keys()), values=list(sc.values()),
                marker_colors=["#dc143c","#ff6400","#ddb200","#00c864"], hole=0.4
            ))
            fig2.update_layout(title="Severity Distribution",
                               paper_bgcolor="#1e2130",font_color="white",height=420)
            st.plotly_chart(fig2, use_container_width=True)

        st.markdown("### Recent Scans")
        tbl = []
        for pat, fjson, date in rows[:20]:
            try:
                fl  = json.loads(fjson)
                top = fl[0]["label"] if fl else "None"
                n   = len(fl)
            except: top,n = "None",0
            tbl.append({"Patient":pat,"Date":date[:16],"Findings":n,"Top Finding":top})
        st.dataframe(pd.DataFrame(tbl), use_container_width=True)
```
